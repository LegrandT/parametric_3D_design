```python
import math
import numpy as np
import matplotlib.pyplot as plt

def generate_cos_curve(amplitude, frequency, phase, num_points):
    """
    Generate points for a cosine curve.

    Parameters:
    amplitude (float): The amplitude of the cosine curve.
    frequency (float): The frequency of the cosine curve.
    phase (float): The phase of the cosine curve.
    num_points (int): The number of points to generate.

    Returns:
    list: A list of tuples representing the generated points in the form (x, y).
    """

    # Generate points for cos curve
    x = list(range(num_points))
    y = [amplitude * np.sin(frequency * x + phase) for x in x]

    return list(zip(x, y))

def generate_sin_curve(amplitude, frequency, phase, num_points):
    """
    Generate points for a sine curve.

    Parameters:
    amplitude (float): The amplitude of the sine curve.
    frequency (float): The frequency of the sine curve.
    phase (float): The phase of the sine curve.
    num_points (int): The number of points to generate.

    Returns:
    list: A list of tuples representing the generated points in the form (x, y).
    """

    # Generate points for sin curve
    x = list(range(num_points))
    y = [amplitude * np.sin(frequency * x + phase) for x in x]

    return list(zip(x, y))

# [2 * math.sin(math.pi * i / num_layers) for i in range(num_layers)]

# Parameters
amplitude = 1 # hight of the curve
period = 100 # width of the curve
frequency = 2 * math.pi / period # frequency of the curve
phase = 0 # shift the curve to the right
num_points = 100

# Generate points for cos curve
points_cos = generate_cos_curve(amplitude, frequency, phase, num_points)
points_sin = generate_sin_curve(amplitude, frequency, phase, num_points)

# Plot cos curve
plt.plot([x for x, y in points_cos], [y for x, y in points_cos], label='cos')
plt.scatter([x for x, y in points_cos], [y for x, y in points_cos])
plt.plot([x for x, y in points_sin], [y for x, y in points_sin], label='sin')
plt.scatter([x for x, y in points_sin], [y for x, y in points_sin])
plt.legend()
plt.show()
```
