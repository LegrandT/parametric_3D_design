```python
import math
import numpy as np
import matplotlib.pyplot as plt

def generate_cos_curve(amplitude, frequency, phase, height, num_points):
    """
    Generate points for a cosine curve.

    Parameters:
    amplitude (float): The amplitude of the cosine curve.
    frequency (float): The frequency of the cosine curve.
    phase (float): The phase of the cosine curve.
    num_points (int): The number of points to generate.

    Returns:
    list: A list of tuples representing the generated points in the form (x, y).
    """

    # Generate points for cos curve
    x = list(range(num_points))
    y = [height+amplitude * np.cos(frequency * x + phase) for x in x]

    return list(zip(x, y))

def generate_sin_curve(amplitude, frequency, phase, height, num_points):
    """
    Generate points for a sine curve.

    Parameters:
    amplitude (float): The amplitude of the sine curve.
    frequency (float): The frequency of the sine curve.
    phase (float): The phase of the sine curve.
    num_points (int): The number of points to generate.

    Returns:
    list: A list of tuples representing the generated points in the form (x, y).
    """

    # Generate points for sin curve
    x = list(range(num_points))
    y = [height+amplitude * np.sin(frequency * x + phase) for x in x]

    return list(zip(x, y))



# [2 * math.sin(math.pi * i / num_layers) for i in range(num_layers)]

# Parameters
amplitude = 10 # hight of the curve
period = 200 # width of the curve
frequency = 2 * math.pi / period # frequency of the curve
phase = 0 # shift the curve to the right
num_points = 200
height = 0

# Generate points for cos curve
points_cos = generate_cos_curve(amplitude, frequency, phase, height,num_points)
point_sin = generate_sin_curve(amplitude, frequency, phase, height,num_points)
print(points_cos)
print(point_sin)

# # calculate the difference between one point and the next
# diff = [0]+[points_cos[i+1][1] - points_cos[i][1] for i in range(len(points_cos)-1)]

curve_sum = [(points_cos[i][0], points_cos[i][1] + point_sin[i][1]) for i in range(len(points_cos))]
print(np.sin(0))
print(np.cos(0))
# Plot cos curve
plt.plot([x for x, y in points_cos], [y for x, y in points_cos], label='cos')
plt.scatter([x for x, y in points_cos], [y for x, y in points_cos])
plt.plot([x for x, y in point_sin], [y for x, y in point_sin], label='sin')
plt.scatter([x for x, y in point_sin], [y for x, y in point_sin])
plt.plot([x for x, y in curve_sum], [y for x, y in curve_sum], label='sum')
plt.legend()
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def get_sin_cos(function='sin', amplitude=1, period=1, phase=0, n_points=100):
        scaled_period = period/1
        if function == 'sin':
            return [amplitude * math.sin(2 * math.pi * i / scaled_period + 2*math.pi*phase/scaled_period) for i in range(n_points)]
        elif function == 'cos':
            return [amplitude * math.cos(2 * math.pi * i / scaled_period + 2*math.pi*phase/scaled_period) for i in range(n_points)]
        else:
            return None


params = {
                "type":"cos",
                "amplitude":5,
                "period":100,
                "phase":50
            }


t1 = get_sin_cos(params['type'], amplitude=params['amplitude'], period=params['period'], phase=params['phase'], n_points=100)


plt.plot(t1)
plt.show()
```
